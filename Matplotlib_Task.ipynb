## Q5. Plot multiple lines in the same chart

```python
days = [1, 2, 3, 4, 5]
max_t = [50, 51, 52, 48, 47]
min_t = [43, 42, 40, 44, 33]
avg_t = [45, 48, 48, 46, 40]
```

```python
plt.xlabel('Day')
plt.ylabel('Temprature')
plt.title('Weather')
plt.plot(days, max_t, label="max")
plt.plot(days, min_t, label="min")
plt.plot(days, avg_t, label="average")
plt.show()
```

## Q10. Plot a pie chart using plt.pie()

```python
labels = ['Apple', 'Banana', 'Cherry', 'Dates']
sizes = [30, 25, 25, 20]  # Percentages
colors = ['red', 'yellow', 'pink', 'brown']
explode = (0.1, 0, 0, 0)  # "Explode" the 1st slice (Apple)

plt.pie(sizes, labels=labels, colors=colors, explode=explode, autopct='%1.1f%%', shadow=True, startangle=140)

plt.title('Fruit Distribution')
plt.axis('equal')
plt.show()
```

# Q11. Use subplots to display multiple plots

```python
x = [1, 2, 3, 4, 5]
y1 = [2, 4, 12, 18, 27]
y2 = [27, 18, 12, 4, 2]

plt.subplot(1, 2, 1)
plt.plot(x, y1, 'r-o')
plt.title('Plot 1')

plt.subplot(1, 2, 2)
plt.plot(x, y1, 'g--s')
plt.title('Plot 2')

plt.tight_layout()
plt.show()
```

## Q14. Plot a scatter plot using plt.scatter()

```python
x = [5, 7, 8, 7, 2, 17, 2, 9, 4, 11]
y = [99, 86, 87, 88, 100, 86, 103, 87, 94, 78]

plt.scatter(x, y, color='purple', marker='x')

plt.title("Sample Scatter Plot")
plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.grid(True)
plt.show()
```

## Q15. Add annotations to plot points

```python
x = [1, 2, 3, 4, 5]
y = [10, 20, 15, 25, 30]

plt.scatter(x, y, color='blue', marker='o')

for i in range(len(x)):
    plt.annotate(f'({x[i]}, {y[i]})', (x[i], y[i]), textcoords="offset points", xytext=(5,5), ha='center')

plt.title("Scatter Plot with Annotations")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.grid(True)
plt.show()
```

## Q16. Create a stacked bar chart

```python
labels = ['A', 'B', 'C', 'D']
group1 = [10, 15, 20, 25]
group2 = [5, 10, 5, 15]

plt.bar(labels, group1, label='Group 1')
plt.bar(labels, group2, bottom=group1, label='Group 2')
plt.xlabel('Category')
plt.ylabel('Value')
plt.title('Stacked Bar Chart')
plt.legend()
plt.show()
```

## Q17. Change background and axis colors

```python
plt.plot([1, 2, 3], [4, 5, 6], color='yellow')

ax = plt.gca()
ax.set_facecolor('black')
ax.tick_params(colors='white')
plt.title("Simple Plot", color='white')
plt.show()
```

## Q18. Create a plot with twin axes (twinx/twiny)

```python
x = [1, 2, 3]
y1 = [1, 4, 9]
y2 = [9, 4, 1]

ax1 = plt.gca()
ax1.plot(x, y1, 'r')
ax2 = ax1.twinx()
ax2.plot(x, y2, 'b')

plt.show()
```
